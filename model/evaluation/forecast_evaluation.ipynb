```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading Dataset, converting 'Month' which is the date to datettime and setting it as index

forecasted_results = pd.read_csv('forecast_results.csv',
                                 parse_dates=['Month'],
                                 index_col='Month')
forecasted_results
```

```python
fig, ax = plt.subplots(figsize=(12,6))

# Plot the actuals and forecasts
ax.plot(forecasted_results['Actuals'], label='Actuals', color='black', linestyle='--', linewidth=3)
ax.plot(forecasted_results['TFT'], label='TFT', color='red', linewidth=3)
ax.plot(forecasted_results['TFT (Without External Regressors)'], label='TFT (Without External Regressors)', color='green', linewidth=3)
ax.plot(forecasted_results['SARIMAX'], label='ARIMAX', color='blue', linewidth=3)
ax.plot(forecasted_results['ARIMAX'], label='SARIMA', color='purple', linewidth=3)
ax.plot(forecasted_results['SARIMA'], label='SARIMA', color='orange', linewidth=3)
ax.plot(forecasted_results['ARIMA'], label='ARIMA', color='orange', linewidth=3)

# Add a title and axis labels
ax.set_title('Comparison of Various Time Series Models')
ax.set_xlabel('Date')
ax.set_ylabel('Forecasted Price')

# Add a legend
ax.legend(loc='upper left')

# Display the plot
plt.show()
```
